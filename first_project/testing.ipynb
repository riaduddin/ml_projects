```python
import numpy
from sklearn.datasets import load_svmlight_file
```

```python


# Load the training set for Fold 1
data_path = "Letor/OHSUMED/Data/Fold1/trainingset.txt"

# query_id=True extracts the 'qid' values into the 'query_id' variable
X, y, query_id = load_svmlight_file(data_path, query_id=True)

print(f"Number of samples: {X.shape[0]}")
print(f"Number of features: {X.shape[1]}")
print(f"Unique Query IDs: {len(set(query_id))}")
```

```python
import pandas as pd

def load_letor_to_dataframe(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            parts = line.split('#')[0].strip().split()
            label = int(parts[0])
            qid = parts[1].split(':')[1]
            features = {p.split(':')[0]: float(p.split(':')[1]) for p in parts[2:]}
            
            row = {'label': label, 'qid': qid}
            row.update(features)
            data.append(row)
            
    return pd.DataFrame(data)

df = load_letor_to_dataframe("Letor/OHSUMED/Data/Fold1/trainingset.txt")
df.head()
```

```python
import matplotlib.pyplot as plt

docs_per_query = df.groupby('qid').size()

plt.hist(docs_per_query, bins=30)
plt.xlabel("Documents per Query")
plt.ylabel("Number of Queries")
plt.title("Docs per Query Distribution")
plt.show()
```

```python
query_relevance = df.groupby('qid')['label'].mean()

plt.hist(query_relevance, bins=30)
plt.title("Average Relevance per Query")
plt.show()
```
